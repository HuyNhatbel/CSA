```python
import pandas as pd
df = pd.read_csv("billionaires.csv")
df.head()
```

# Xử lí dữ liệu NaN
- isnull().sum()
- dropna(axis=1)
- drop(['ten cot'], axis=1, inplace=True): 'bỏ các cột không cần thiết

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
# Hàm ghi đè lên dữ liệu cũ
df.drop(["organization"], axis=1, inplace=True)
```

# Điền khuyết
- fillna()
- bfill: `thay giá trị khuyết = giá trị ngay sau nó`
- ffill: `thay giá trị khuyết = giá trị ngay trước nó`

```python
df_filled_0 = df.fillna(0)
df_filled_0.isnull().sum()
df_filled_0.tail()
```

```python
df_bfill = df.bfill()
df_bfill.isnull().sum()
```

```python
df_ffill = df.ffill()
df_ffill.isnull().sum()
```

# Ép kiểu
- dtypes
- Series.astype('<kiểu dữ liệu>')
- pd.to_numeric(<cột>): `Để panda tự chọn kiểu phù hợp`

```python
df.dtypes
```

```python
df_ffill.age = pd.to_numeric(df_ffill.age)
df_ffill.age.dtype
```

```python
df_ffill.birthYear = df_bfill.birthYear.astype("int")
df_ffill.birthYear.dtype
```
