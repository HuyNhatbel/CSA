```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("hw7_cars.csv")
df.tail()
```

Câu 1: Trong dữ liệu có: giá xe, số dặm đi được, tuổi, năm và mẫu xe
Những đặc điểm này có tương quan với giá xe

```python
df.isnull().sum()
```

```python
model = LinearRegression()
X = df[["year"]]
y = df["Price"]
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
print("Hệ số hồi quy (slope):", model.coef_[0])
print("Hệ số chặn (intercept):", model.intercept_)
print(f"Phương trình: year = {model.intercept_:.2f} + {model.coef_[0]:.2f} * Price")
```

```python
plt.figure(figsize=(8,6))
sb.scatterplot(x=df["year"], y=df["Price"], color="red")

plt.title("Biểu đồ thể hiện mối quan hệ giữa năm sản xuất và giá xe")
plt.xlabel("Năm sản xuất")
plt.ylabel("Giá xe ($)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sb.scatterplot(x=X["year"], y=y, color="red", label="Dữ liệu thực tế")
plt.plot(X["year"], y_pred, color="orange", linewidth=2, label="Đường hồi quy tuyến tính")
plt.title("Mối quan hệ giữa năm sản xuất và giá thành xe")
plt.xlabel("Năm sản xuất")
plt.ylabel("Giá thành xe ($)")
plt.legend()
plt.show()
```
